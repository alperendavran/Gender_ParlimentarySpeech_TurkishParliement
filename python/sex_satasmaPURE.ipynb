```python
import pandas as pd
import matplotlib.pyplot as plt

party_gender_colors = {
    'AK Parti_erkek': 'darkorange',
    'AK Parti_kadin': 'lightsalmon',
    'CHP_erkek': 'darkred',
    'CHP_kadin': '#FF6666',
    'HDP_erkek': 'purple',
    'HDP_kadin': 'violet',
    'MHP_erkek': 'black',
    'MHP_kadin': 'grey',
}

def plot_data():
    terms = [24, 26, 27]
    width = 0.35
    fig, ax = plt.subplots(figsize=(12, 7))
    
    positions = []
    labels = []

    for idx, term in enumerate(terms):
        # men
        bottoms = 0
        for party_gender, color in party_gender_colors.items():
            if '_erkek' in party_gender:
                party_name = party_gender.split('_')[0]  # Extract party name
                count = grouped_data[(grouped_data['PARTY'] == party_name) & (grouped_data['SEX'] == 'erkek') & (grouped_data['TERM'] == term)]['percentage'].values[0]
                ax.bar(idx*3, count, width, bottom=bottoms, color=color)
                ax.text(idx*3, bottoms + count/2, f"{count:.2%}", ha='center', va='center', fontsize=10, color='white')
                bottoms += count
        
        positions.append(idx*3)
        labels.append(f"{term} Male")
        
        # women
        bottoms = 0
        for party_gender, color in party_gender_colors.items():
            if '_kadin' in party_gender:
                party_name = party_gender.split('_')[0]  # Extract party name
                count = grouped_data[(grouped_data['PARTY'] == party_name) & (grouped_data['SEX'] == 'kadin') & (grouped_data['TERM'] == term)]['percentage'].values[0]
                ax.bar(idx*3 + 1, count, width, bottom=bottoms, color=color)
                ax.text(idx*3 + 1, bottoms + count/2, f"{count:.2%}", ha='center', va='center', fontsize=10, color='white')
                bottoms += count

        positions.append(idx*3 + 1)
        labels.append(f"{term} Female")

    # Manually constructing the legend
    from matplotlib.patches import Patch
    legend_handles = [Patch(color=value, label=key.replace('_', ' ')) for key, value in party_gender_colors.items()]
    ax.legend(handles=legend_handles, loc='upper right')
    
    ax.set_title('Aggression in parliament based on sex and party across Terms 24, 26, 27 [Excluding Head MPs]')
    ax.set_ylabel('Percentage of teasings')
    ax.set_xticks(positions)
    ax.set_xticklabels(labels)
    plt.tight_layout()
    plt.show()

plot_data()
```
