```python
def combined_value_counts(df, prefix="commission_", start=1, end=6):
    value_counts_list = []
    
    for i in range(start, end+1):
        col_name = f"{prefix}{i}"
        if col_name in df.columns:
            value_counts_list.append(df[col_name].value_counts())
    
    combined = pd.concat(value_counts_list, axis=1).fillna(0)    
    return combined

# Sample usage
m=combined_value_counts(cv)
```
