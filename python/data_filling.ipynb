```python
def clean_text(speech_text):
    extracted_text = speech_text.strip()
    words = extracted_text.split()  # Split the string into words
    extracted_text = ' '.join(words)  # Words back together with single spaces
    return extracted_text


df['cv'] = df['cv'].apply(clean_text)

df.to_excel('cv_27.xlsx', index=False)  
```

```python
import pandas as pd
import re
from TurkishStemmer import TurkishStemmer

stemmer = TurkishStemmer()

places_df = pd.read_csv("/Users/alperendavran/Desktop/Gender and Parlimentary Speech/trash /il-ilçe.csv", sep=';')
places_df = places_df[places_df['il'].notna() & places_df['ilçe'].notna()]

places_df['il'] = places_df['il'].apply(lambda x : str(x))
places_df['ilçe'] = places_df['ilçe'].apply(lambda x : str(x))

cities = places_df['il'].unique().tolist()
districts = places_df['ilçe'].unique().tolist()

cities = [each.strip() for each in cities]
districts=[each.strip() for each in districts]

for i in range(len(cities)):
    cities[i] = cities[i].replace("İ", "I").replace("ı", "I")
    cities[i] = cities[i].replace("Ü", "U").replace("ü", "u")
    cities[i] = cities[i].replace("Ö", "O").replace("ö", "o")

for i in range(len(districts)):
    districts[i] = districts[i].replace("İ", "I").replace("ı", "I")
    districts[i] = districts[i].replace("Ü", "U").replace("ü", "u")
    districts[i] = districts[i].replace("Ö", "O").replace("ö", "o")




```

```python
import pandas as pd

def remove_after_apostrophe(text):
    text=text.upper()
    words = text.split()
    for i, word in enumerate(words):
        if "'" in word:
            words[i] = word.split("'")[0]
    return " ".join(words)


for i, row in df.iterrows():
    cv = row['cv']  
    text = cv[0:cv.find(".")] 
    text= remove_after_apostrophe(text)
    bio_processed = re.sub(r'[^\w\s]', '', text)
    bio_processed = " ".join([word.split("'")[0] for word in bio_processed.split()])
    city = district = 'NONE'

    for word in bio_processed.split():
        if word in cities:
            city = word
            break

    for word in bio_processed.split():
        if word in districts:
          district = word
          break
    if district == 'NONE':
        print(f"Uyarı: {row['vekil_name']} biyografisinde bir ilçe bulunamadı.")
    else:
        print(f"Uyarı: {row['vekil_name']} biyografisinde bir il bulunamadı.")

    df.at[i, 'birth_province'] = city
    df.at[i, 'birth_district'] = district

# DataFrame'i diske yazın
df.to_csv('updated_cv26_bio.csv', index=False)  
```

```python
from datetime import datetime
import pandas as pd
import re

# Türkçe'den İngilizce'ye ay çeviri sözlüğü
aylar_tr_en = {
    "Ocak": "January",
    "Şubat": "February",
    "Mart": "March",
    "Nisan": "April",
    "Mayıs": "May",
    "Haziran": "June",
    "Temmuz": "July",
    "Ağustos": "August",
    "Eylül": "September",
    "Ekim": "October",
    "Kasım": "November",
    "Aralık": "December"
}


for i, row in df.iterrows():
    cv = row['cv']
    text = cv[0:cv.find(".")]  # cv'nin ilk cümlesini al
    match = re.search(r"(?:(\d+)\s+)?(?:(\w+)\s+)?(\d{4})", text)
    if match:
        groups = match.groups()

        gun = groups[0] or 'NONE'
        ay = aylar_tr_en.get(groups[1], 'NONE') if groups[1] else 'NONE'
        yil = groups[2] or 'NONE'

        # Tarihi ayrıştır
        if gun != 'NONE' and ay != 'NONE' and yil != 'NONE':
            try:
                tarih = datetime.strptime(f"{gun} {ay} {yil}", "%d %B %Y")
                df.at[i, "birth_year(yyyy)"] = tarih.year
                df.at[i, "birth_month(1-12)"] = tarih.month
                df.at[i, "birth_day(1-31)"] = tarih.day
            except ValueError:
                df.at[i, "birth_year(yyyy)"] = 'NONE'
                df.at[i, "birth_month(1-12)"] = 'NONE'
                df.at[i, "birth_day(1-31)"] = 'NONE'
        elif ay != 'NONE' and yil != 'NONE':
            try:
                tarih = datetime.strptime(f"{ay} {yil}", "%B %Y")
                df.at[i, "birth_year(yyyy)"] = tarih.year
                df.at[i, "birth_month(1-12)"] = tarih.month
                df.at[i, "birth_day(1-31)"] = 'NONE'
            except ValueError:
                df.at[i, "birth_year(yyyy)"] = 'NONE'
                df.at[i, "birth_month(1-12)"] = 'NONE'
                df.at[i, "birth_day(1-31)"] = 'NONE'
        elif yil != 'NONE':
            df.at[i, "birth_year(yyyy)"] = int(yil)
            df.at[i, "birth_month(1-12)"] = 'NONE'
            df.at[i, "birth_day(1-31)"] = 'NONE'
        else:
            df.at[i, "birth_year(yyyy)"] = 'NONE'
            df.at[i, "birth_month(1-12)"] = 'NONE'
            df.at[i, "birth_day(1-31)"] = 'NONE'
    else:
        df.at[i, "birth_year(yyyy)"] = 'NONE'
        df.at[i, "birth_month(1-12)"] = 'NONE'
        df.at[i, "birth_day(1-31)"] = 'NONE'


df.to_csv('2hfilled_cv_27period.csv', index=False)  
```

```python
import requests
import json

def getGenders(names):
	url = ""
	cnt = 0
	if not isinstance(names,list):
		names = [names,]

	for name in names:
		if url == "":
			url = "name[0]=" + name
		else:
			cnt += 1
			url = url + "&name[" + str(cnt) + "]=" + name


	req = requests.get("https://api.genderize.io?" + url)
	results = json.loads(req.text)

	retrn = []
	for result in results:
		if result["gender"] is not None:
			retrn.append((result["gender"], result["probability"], result["count"]))
		else:
			retrn.append((u'None',u'0.0',0.0))
	return retrn


for i, row in df.iterrows():
    name = row['vekil_name']
    name = name[0:name.find("_")]
    genderList = getGenders([name])
    gender = genderList[0]  #first indeces
    if float(gender[1]) > 0.8:  # Cinsiyet olasılığı
        if gender[0] == 'male':
            df.loc[i, 'sex'] = "erkek"
        else:
            df.loc[i, 'sex'] = "kadın"
    else:
       df.loc[i, 'sex'] = "NONE"
df.to_csv('updated3_cv27_bio.csv', index=False) 
```

```python
def find_second_last_dot(s):
    last_dot_index = s.rfind('.')
    if last_dot_index == -1:  # Hiç nokta bulunmamışsa
        return -1
    second_last_dot_index = s.rfind('.', 0, last_dot_index)
    return second_last_dot_index

for i, row in df.iterrows():
    cv = row['cv']
    text = cv[find_second_last_dot(cv)+1:]
    list_text = text.split()
    if "çocuk" in list_text:  # 'çocuk' kelimesi metinde bulunuyor mu diye kontrol et
        index = list_text.index("çocuk")  # 'çocuk' kelimesinin indexini bul
        df.loc[i, 'kids'] = list_text[index-1] # 'çocuk' kelimesinden önce gelen kelimeyi yazdır
    else:
        df.loc[i, 'kids'] = 0
df.to_csv('updated5_cv27_bio.csv', index=False)  # 'guncellenmis_biyografiler.csv' yerine istediğiniz dosya adını kullanabilirsiniz
```

```python
from transformers import AutoModelForTokenClassification
from transformers import pipeline
from transformers import AutoModel, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("dbmdz/bert-base-turkish-cased")
model = AutoModel.from_pretrained("dbmdz/bert-base-turkish-cased")

model = AutoModelForTokenClassification.from_pretrained("akdeniz27/bert-base-turkish-cased-ner")
tokenizer = AutoTokenizer.from_pretrained("akdeniz27/bert-base-turkish-cased-ner")
ner = pipeline('ner', model=model, tokenizer=tokenizer, aggregation_strategy="first")
```
